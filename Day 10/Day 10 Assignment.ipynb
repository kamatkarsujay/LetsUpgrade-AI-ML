```python
# Import all the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data into data1 variable
data1 = pd.read_csv('general_data.csv')
data1.head()
```

```python
# Get the no of rows and columns in the data
data1.shape
```

```python
# Get the information about each column
data1.info()
```

```python
# Count the no of null values
data1.isna().sum()
```

```python
# Drop the null values
data1.dropna()
```

```python
# Get the duplicate values
data1.duplicated()
```

```python
# Drop the duplicate values
data1.drop_duplicates()
```

```python
# Get some basic statistical values
data1.describe()
```

```python
# Median of each column
data1.median()
```

```python
# Mode of each column
data1.mode()
```

```python
# Variance of each column
data1.var()
```

```python
# Skewness of each column
data1.skew()
```

```python
# Kurtosis of each column
data1.kurt()
```

```python
# Inter Quartile range
Q1 = np.percentile(data1.MonthlyIncome, 25)
Q3 = np.percentile(data1.MonthlyIncome, 75)
print('IQR',Q3-Q1)
```
